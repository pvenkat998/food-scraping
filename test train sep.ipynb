```python
filepath = 'train_list.txt'
with open(filepath) as fp:
    line = fp.readline()
    trainlist=[]
    while line:
       # print("Line {}: {}".format(cnt, line.strip()))
        trainlist.append(line.strip()[7:])
        line = fp.readline()
print(len(trainlist))
```

```python
filepath = 'test_list.txt'
with open(filepath) as fp:
    line = fp.readline()
    testlist=[]
    while line:
       # print("Line {}: {}".format(cnt, line.strip()))
        testlist.append(line.strip()[7:])
        line = fp.readline()
print(len(testlist))
```

```python
import shutil
import os
cur_dir = os.getcwd() # current dir path
cur_dir=os.path.join(cur_dir, 'data')
list_dir = os.listdir(cur_dir)
print(cur_dir)
trc=0
tec=0
desttr = os.path.join(cur_dir,'train') 
destte = os.path.join(cur_dir,'test') 
print(desttr)
print(destte)
for sub_dir in list_dir:
    if sub_dir in trainlist:
        print("train")
        dir_to_move = os.path.join(cur_dir, sub_dir)
        print(dir_to_move)
        shutil.move(dir_to_move, desttr)
        trc=trc+1
    elif sub_dir in testlist:
        dir_to_move = os.path.join(cur_dir, sub_dir)
        shutil.move(dir_to_move, destte)
        print("test")
        tec=tec+1
    else:
        print("wut")
print(trc)
print(tec)
```
